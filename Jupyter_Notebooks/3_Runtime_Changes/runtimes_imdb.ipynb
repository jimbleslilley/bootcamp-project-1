# How runtime affects the scores on an IMDB

```python
#import dependecies
import pandas as pd
import matplotlib as plt
```

```python
#read the csv into dataframe
net_df = pd.read_csv('../../Resources/netflix_omdb.csv',index_col=0)
```

```python
#display dataframe
net_df
```

## Clean the dataframe

```python
# find the null values in dataframe
null_df = net_df[net_df.isna().any(axis=1)]
null_df
```

```python
#dropping the null values
new_net_df = (net_df.dropna(how= 'any')).reset_index(drop = True)
new_net_df
```

```python
# sorting the dataframe on runtime 
new_net_df.sort_values(by=['Runtime'])
```

```python
new_net_df.to_csv('../../Resources/clean_netflix_shows.csv')
```
